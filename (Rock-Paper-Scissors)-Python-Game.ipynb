```python
#Abhishek Gupta

import random

def user_choice():
    a=["Rock","rock","Paper","paper","Scissor","scissor"]
    i=input(f" Player Enter your Move :: [rock / paper / scissor] ").lower()
    while i not in a:
        print(f"Invalid Choice  ! Try Again..")
        i=input(f"  Enter your Move :: [ROCK / PAPER / SCISSOR] ").lower()
    return i

def computer_random():
    b=random.choice(["rock","paper","scissor"])
    return b

def r(user,comp):
    if user==comp:
        return"--DRAW--"
    elif user == "rock" and comp == "scissor":
        return"You Win "
    elif user == "paper" and comp == "rock":
        return"You Win "
    elif user == "scissor" and comp == "paper":
        return"You Win "
    else:
        return"**Computer Wins**"
        
n=input("--Enter Your Name :")
print(f"Hello {n} \n\nWelcome To **Rock Paper Scissor Game**")

while True:
    
    user=user_choice()
    comp=computer_random()
    print(f"\nYou chose: {user}")
    print(f"Computer chose: {comp}")
    result= r(user,comp)
    print("------------------------------------------RESULT----------------------------------------------------")
    print("Result:", result)

    again=input("Do You Want To Play Again (Yes/No) ? ").lower()
    if again != "yes":
        print(f"Bye Bye {n} ! Thankyou For Playing....")
        break
```
